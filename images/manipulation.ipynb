# Image manipulation

## Extract images from video

```python
# Importing all necessary libraries
import cv2
import os
  
# Read the video from specified path
cam = cv2.VideoCapture("/Users/sergiocuencanunez/Downloads/video5.mp4")
  
try:
      
    # creating a folder named data
    if not os.path.exists('data'):
        os.makedirs('data')
  
# if not created then raise errorx
except OSError:
    print ('Error: Creating directory of data')
  
# frame
currentframe = 0
  
while(True):
      
    # reading from frame
    ret,frame = cam.read()
  
    if ret:
        # if video is still left continue creating images
        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)
  
        # writing the extracted images
        cv2.imwrite(name, frame)
  
        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break
  
# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

## Crop and Resize images extracted

```python
import cv2
import os

input_images_path = "/Users/sergiocuencanunez/Downloads/Sergio_Cuenca_input2"
files_names = os.listdir(input_images_path)
print(files_names)

output_images_path = "/Users/sergiocuencanunez/Downloads/Sergio_Cuenca_TEST"
if not os.path.exists(output_images_path):
    os.makedirs(output_images_path)
    print("Directorio creado: ", output_images_path)

count = 0
for file_name in files_names:

    image_path = input_images_path + "/" + file_name
    image = cv2.imread(image_path)
    if image is None:
        continue
    
    image = image[50:750,450:1150]
    image = cv2.resize(image, (250, 250))

    cv2.imwrite(output_images_path + "/Sergio_Cuenca_000" + str(int(count+1)) + ".jpg", image)
    count += 1
```
